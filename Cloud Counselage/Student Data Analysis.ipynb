## 3.What is the distribution of students across different graduation years?

```python
graduation_year_distribution = data['Year of Graduation'].value_counts().sort_index()
# dataframe for data distribution 
distribution = pd.DataFrame({
    'year of graduation': graduation_year_distribution.index,
    'number of students': graduation_year_distribution.values})
# Barplot
plt.figure(figsize=(5, 3))
sns.barplot(x='year of graduation', y='number of students', data=distribution, palette='Set1')
plt.title('Distribution of students across graduation years',size=15,color='black')
plt.xlabel('Year of graduation',size=10,color='black')
plt.ylabel('Number of students',size=10,color='black')
plt.tight_layout()
for i, value in enumerate(graduation_year_distribution.values):
    plt.text(i, value + 1, str(value), ha='center', va='center',size=10,color='black')
plt.show()
```

## 4.What is the distribution of student's experience with Python programming?

```python
# Calculating the distribution of students and experience with Python programming
experience_distribution = data['Experience with python (Months)'].value_counts().sort_values(ascending=True)
#Plotting
plt.figure(figsize=(5, 3))
plt.bar(experience_distribution.index, experience_distribution.values, color='blue', edgecolor='black')
plt.title("Distribution of Students' Experience with Python Programming",size=15,color='black')
plt.xlabel('Months of Experience',size=11,color='black')
plt.xticks(size=12,color='black')
plt.ylabel('Number of Students',size=11,color='black')
plt.yticks(size=12,color='black')
plt.tight_layout()
plt.show()
```

## 5.What is the average family income of the student?

## 6.How does the GPA vary among different colleges? (Show top 5 results only)

```python
# Calculating GPA averages for different colleges with  top 5 results
college_gpa_averages = data.groupby('College Name')['CGPA'].mean()
top_colleges_gpa = college_gpa_averages.nlargest(5).sort_values(ascending=True)
#Ploting
plt.barh(top_colleges_gpa.index, top_colleges_gpa,color='grey')
plt.xlabel('Average GPA',size=14,color='black')
plt.xticks(size=12,color='black')
plt.ylabel('College Name',size=14,color='black')
plt.yticks(size=12,color='black')
plt.title('Top 5 Colleges with Highest Average GPA',size=15,color='black')
plt.tight_layout()
plt.show()
```

## 7.What is the average GPA for student from each city?

```python
# Calculating average GPA for each city
city_gpa_averages = data.groupby('City')['CGPA'].mean()
# Sorting the cities based on average GPA in descending order
sorted_city_gpa = city_gpa_averages.sort_values(ascending=False)
# horizontal bar plot
plt.figure(figsize=(10, 30))
ax = sns.barplot(x=sorted_city_gpa, y=sorted_city_gpa.index, palette='viridis')
plt.title("Average GPA for Students from Each City",size=15,color='black')
plt.xlabel("Average GPA",size=12,color='black')
plt.xticks(size=12,color='black')
plt.ylabel("City",size=12,color='black')
plt.yticks(size=12,color='black')
for p in ax.patches:
    ax.annotate(f"{p.get_width():.2f}", (p.get_width(), p.get_y() + p.get_height() / 2.), va='center', xytext=(5, 0), textcoords='offset points', color='black', size=10)
plt.tight_layout()
plt.show()
```

## 8.Can we identify any relationship between family income and GPA?

```python
# Scatter plot with regression line
plt.figure(figsize=(6, 4))
plt.scatter(data["Family Income"], data["CGPA"], alpha=0.5, label="Data Points",color='green')
# Calculating the regression line
slope, intercept, r_value, p_value, std_err = linregress(data["Family Income"], data["CGPA"])
regression_line = slope * data["Family Income"] + intercept
#plotting
plt.plot(data["Family Income"], regression_line, color="red", label="Regression Line")
plt.title("Family Income vs CGPA with Regression Line",size=15,color='black')
plt.xlabel("Family Income",size=12,color='black')
plt.xticks(size=12,color='black')
plt.ylabel("CGPA",size=12,color='black')
plt.yticks(size=12,color='black')
plt.legend()
plt.show()
```

## 15.Those who have high CGPA & More experience in language those who had high expectations for Salary? (Avg)

```python
# Define high CGPA as greater than or equal to a certain threshold (e.g., 3.5)
high_cgpa_threshold = 3.5
# Filter the data for students with high CGPA and more experience in the language
high_cgpa_language = data[(data['CGPA'] >= high_cgpa_threshold) & (data['Experience with python (Months)'] > 12)]
# Calculate the average expected salary for the selected group
average_salary = high_cgpa_language['Expected salary (Lac)'].mean()
print("Average expected salary for students with high CGPA and more experience in the language:", average_salary)
```
